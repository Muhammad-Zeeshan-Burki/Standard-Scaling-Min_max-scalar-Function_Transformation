```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_excel("For encoding prac.xlsx")
dataset.head(3)
```

```python
sns.displot(dataset["Unique.Visits"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["Unique.Visits"]])
```

```python
ss.transform(dataset[["Unique.Visits"]])
```

```python
dataset["Unique.Visits_SS"]=pd.DataFrame(ss.transform(dataset[["Unique.Visits"]]),columns=["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(5,8))

plt.subplot(1,2,1)
plt.title("Before")
sns.displot("Unique.Visits")

plt.subplot(1,2,2)
plt.title("After")
sns.subplot(dataset["Unique.Visits_SS"])
plt.show()
```
