```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel("For encoding prac.xlsx")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
sns.distplot(dataset["Returning.Visits"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MMS =MinMaxScaler()
MMS.fit(dataset[["Returning.Visits"]])
```

```python
dataset["Returning.Visits_MMS"]=MMS.transform(dataset[["Returning.Visits"]])


```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(5,8))
plt.subplot(1,2,1)
plt.title("Before")
sns.displot(dataset["Returning.Visits"])
plt.subplot(1,2,2)
plt.title("After")
sns.displot(dataset["Returning.Visits_MMS"])
plt.show()
```
