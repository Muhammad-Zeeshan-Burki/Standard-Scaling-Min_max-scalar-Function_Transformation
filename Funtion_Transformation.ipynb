```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
```

```python
dataset= pd.read_excel("For encoding prac.xlsx")
dataset.head(5)
```

```python
sns.distplot(dataset["Unique.Visits"])
plt.show()
```

```python
q1 = dataset["Unique.Visits"].quantile(0.25)
q3= dataset["Unique.Visits"].quantile(0.75)
IQR=q3-q1
```

```python
min_value = q1-(1.5*IQR)
max_Value = q3+(1.5*IQR)
min_value,max_Value
```

```python
dataset =dataset[dataset["Unique.Visits"]<=max_Value]
```

```python
sns.distplot(dataset["Unique.Visits"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
```

```python
ft.fit(dataset[["Unique.Visits"]])

```

```python
dataset["Unique.Visits_TF"]=ft.transform(dataset[["Unique.Visits"]])
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.displot(dataset["Unique.Visits"])
plt.title("Before")

plt.subplot(1,2,1)
sns.displot(dataset["Unique.Visits_TF"])
plt.title("After")
plt.show()

```
